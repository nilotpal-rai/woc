```python
import numpy as np
theta = np.zeros((785,1))
def GradientDescent(train_x, train_y, alpha, iterations):
        m = train_y.size #number_of_training_examples
        cost = [] #emptylist_costfunction
        for i in range(0,iterations):
            pred_y = np.dot(train_x, theta) #predicted_value_matrix_multiplication
            cost_func = (1/(2*m))*np.sum((pred_y-train_y)**2)
            d_theta = (1/m)*np.dot(train_x.T, pred_y-train_y) #differentiate_cost_function
            
            theta = theta-alpha*d_theta #updation
            cost.append(cost_func)
        return cost,theta
        cost, theta = GradientDescent(train_x, train_y, 0.01, 1500)
```
